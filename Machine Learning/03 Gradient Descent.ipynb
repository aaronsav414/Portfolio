## Slope & Derivatives
### Create a python function for the derivative of f(x) called df(x)

```python
def df(x):
    return 2*x + 1
```

## Python Loops & Gradient Descent

```python
# For Loop
for n in range(5):
    print("Hello World", n)
print("End of loop")
```

```python
# While Loop
counter = 0
while counter < 7:
    print("Counting...", counter)
    counter += 1
print("Ready or not, here i come.")
```

```python
# Gradient Descent
new_x = 3
prev_x = 0
step_multiplier = 0.1
precision = 0.00000001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    prev_x = new_x
    gradient = df(prev_x)
    new_x = prev_x - step_multiplier * gradient
    
    step_size = abs(new_x - prev_x)
#     print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if step_size < precision:
        print("loop ran this many times:", n)
        break
    
print("Local minimum occurs at:", new_x)
print("Slope or df(x) value at this point is:", df(new_x))
print("f(x) value or cost at this point is:", f(new_x))
```

```python
# Super impose the gradient descent calculations
# First Chart Cost Function
pyp.figure(figsize=[10, 10])

pyp.subplot(2, 1, 1)

pyp.xlim([-3, 3])
pyp.ylim(0, 8)
pyp.title("Cost Function", fontsize=17)
pyp.xlabel('X', fontsize=16)
pyp.ylabel("f(x)", fontsize=16)
pyp.plot(x_1, f(x_1), color="blue", linewidth=3, alpha=0.5)

values = np.array(x_list)
pyp.scatter(x_list, f(values), color="red", s=100, alpha=0.6)

# Second Chart Derivative Function
pyp.subplot(2, 1, 2)

pyp.title("Slopes of the Cost Function", fontsize=17)
pyp.xlabel('X', fontsize=16)
pyp.ylabel("d(x)", fontsize=16)
pyp.grid()
pyp.xlim([-2, 3])
pyp.ylim(-3, 6)

pyp.plot(x_1, df(x_1), color="skyblue", linewidth=6, alpha=0.5)
pyp.scatter(x_list, slope_list, color='red', s=100, alpha=0.6)

pyp.show
```
