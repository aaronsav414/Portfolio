```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as pyp

from sklearn.linear_model import LinearRegression as lr
```

```python
data = pandas.read_csv('/Users/aaronsavary/Documents/Coding/Machine Learning/Excell Table Data Sheets/cost_revenue_clean.csv')
```

```python
data
```

```python
data.describe()
```

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
pyp.figure(figsize=(10, 6))
pyp.scatter(X, y, alpha=0.2)
pyp.title('Film Cost vs. Global Revenue')
pyp.xlabel('Budget $') #feature
pyp.ylabel('Gross $')  #target
pyp.ylim(0, 3000000000)
pyp.xlim(0, 450000000)
pyp.show()
```

```python
r = lr()
r.fit(X, y)
```

Slope Coefficient

```python
r.coef_ # Î¸â‚
```

```python
# Intercept Î¸â‚€
r.intercept_
```

```python
pyp.figure(figsize=(10, 6))
pyp.scatter(X, y, alpha=0.2)
pyp.plot(X, r.predict(X), color='red', linewidth=4)

pyp.title('Film Cost vs. Global Revenue')
pyp.xlabel('Budget $') #feature
pyp.ylabel('Gross $')  #target
pyp.ylim(0, 3000000000)
pyp.xlim(0, 450000000)
pyp.show()
```
